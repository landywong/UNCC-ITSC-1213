# The programs

Below we have the coding project for this demonstration, written over several code blocks.

## Entity

`Entity.java` contains the **superclass** for all characters in the game (e.g., `Player`, `Monster`).

### Key Concepts

- Encapsulation: Uses protected fields.
- Methods: `takeDamage()`, `heal()`, `getName()`, and `toString()` define common behaviors.
- Reusability: Other classes will inherit from this.

````python
%%writefile Entity.java
/**
 * The Entity class is the base (superclass) for all game characters.
 * It includes fields and methods that are common to all entities, such as health and name.
 */
public class Entity {
    protected String name;
    protected int health;
    protected int maxHealth;

    public Entity(String name, int maxHealth) {
        this.name = name;
        this.maxHealth = maxHealth;
        this.health = maxHealth; // Start at full health
    }

    public void takeDamage(int amount) {
        health -= amount;
        if (health < 0) {
            health = 0;
        }
        System.out.println(name + " took " + amount + " damage. Remaining health: " + health);
    }

    public void heal(int amount) {
        health += amount;
        if (health > maxHealth) {
            health = maxHealth;
        }
        System.out.println(name + " healed " + amount + " points. Current health: " + health);
    }

    public String getName() {
        return name;
    }

    public int getHealth() {
        return health;
    }

    @Override
    public String toString() {
        return "Entity{name='" + name + "', health=" + health + "/" + maxHealth + "}";
    }
}
````

## Monster

This class adds a twist: **undead monsters** behave differently when damaged or healed.

### Key Mechanics

- **Undead monsters:**
  - Take half damage.
  - Are harmed by healing (inverts `heal()` method).
- Demonstrates **method overriding**.

````python
%%writefile Monster.java
/**
 * The Monster class extends Entity and demonstrates method overriding.
 * Undead monsters take half damage and are hurt by healing.
 */
public class Monster extends Entity {
    private boolean isUndead;

    public Monster(String name, int maxHealth, boolean isUndead) {
        super(name, maxHealth);
        this.isUndead = isUndead;
    }

    @Override
    public void takeDamage(int amount) {
        if (isUndead) {
            amount /= 2;
            System.out.println(name + " is undead and resists damage!");
        }
        super.takeDamage(amount);
    }

    @Override
    public void heal(int amount) {
        if (isUndead) {
            System.out.println(name + " is undead and is hurt by healing!");
            takeDamage(amount); // Reuse takeDamage to simulate damage from healing
        } else {
            super.heal(amount);
        }
    }

    @Override
    public String toString() {
        return "Monster{name='" + name + "', undead=" + isUndead + ", health=" + health + "/" + maxHealth + "}";
    }
}
````

## Player

This class adds RPG-style mechanics: experience and leveling.

### Key Features

- Adds `experiencePoints` and `level`.
- Gains XP and levels up when threshold is reached.
- Overrides `toString()` for player-specific output.

````python
%%writefile Player.java
/**
 * The Player class extends Entity and adds experience and leveling functionality.
 * It demonstrates inheritance and method overriding.
 */
public class Player extends Entity {
    private int experiencePoints;
    private int level;

    public Player(String name, int maxHealth) {
        super(name, maxHealth); // Calls the constructor of Entity
        this.experiencePoints = 0;
        this.level = 1;
    }

    public void gainExperience(int points) {
        experiencePoints += points;
        System.out.println(name + " gained " + points + " XP!");
        if (experiencePoints >= level * 100) {
            levelUp();
        }
    }

    private void levelUp() {
        level++;
        maxHealth += 10;
        health = maxHealth;
        experiencePoints = 0;
        System.out.println(name + " leveled up! New level: " + level);
    }

    public void printStatus() {
        System.out.println("Player: " + name + ", Level: " + level + ", Health: " + health + "/" + maxHealth + ", XP: " + experiencePoints);
    }

    @Override
    public String toString() {
        return "Player{name='" + name + "', level=" + level + ", health=" + health + "/" + maxHealth + ", XP=" + experiencePoints + "}";
    }
}
````

## Main

This class uses the above objects polymorphically and tests their behaviors.

### What it does

1.	Creates one player and two monsters.
2.	Applies damage to all.
3.	Heals all — undead take damage instead!
4.	Casts the `Entity` back to `Player` to call subclass-specific methods (`gainExperience`, `printStatus`).

````python
%%writefile Main.java
/**
 * Main class to test the inheritance and polymorphism concepts.
 * Demonstrates use of Entity, Player, and Monster in a polymorphic way.
 */
public class Main {
    public static void main(String[] args) {
        Entity player = new Player("Hero", 100);
        Entity goblin = new Monster("Goblin", 80, false);
        Entity zombie = new Monster("Zombie", 120, true);

        Entity[] entities = { player, goblin, zombie };

        // Print initial state
        for (Entity e : entities) {
            System.out.println(e);
        }

        System.out.println("\n--- Applying damage ---");
        for (Entity e : entities) {
            e.takeDamage(20);
        }

        System.out.println("\n--- Healing all entities ---");
        for (Entity e : entities) {
            e.heal(15);
        }

        // Cast player back to Player to gain experience and show specific behavior
        if (player instanceof Player) {
            ((Player) player).gainExperience(120);
            ((Player) player).printStatus();
        }
    }
}
````

## Summary of OOP Conpects in Java Code

| OOP Concept          | Description                                                                 | Example from Code                                                   |
|----------------------|-----------------------------------------------------------------------------|----------------------------------------------------------------------|
| **Class**            | Blueprint for creating objects with fields and methods                      | `Entity`, `Player`, `Monster`                                       |
| **Object**           | Instance of a class                                                         | `Entity player = new Player("Hero", 100);`                          |
| **Inheritance**      | Mechanism where a class inherits behavior and properties from a superclass  | `Player extends Entity`, `Monster extends Entity`                   |
| **Polymorphism**     | Ability to treat different subclasses as if they were the same superclass    | `Entity[] entities = { player, goblin, zombie };`                   |
| **Method Overriding**| Subclasses redefine methods from the superclass                              | `takeDamage()` and `heal()` in `Monster`, `toString()` in both     |
| **Encapsulation**    | Restricting direct access to fields; using methods to interact               | Fields like `health` and `maxHealth` are `protected/private`       |
| **Constructors**     | Special methods to initialize objects                                        | `new Player(...)`, `new Monster(...)`                              |
| **Type Casting**     | Accessing subclass-specific methods from a superclass reference              | `((Player) player).gainExperience(120);`                           |
| **Instance Checking**| Checking object type at runtime                                              | `if (player instanceof Player)`                                    |

### Additional Notes

- **Undead logic** in `Monster` shows real-world behavior simulation via method overrides.
- **Experience and level-up system** in `Player` demonstrates how subclasses add new features.
- `super(...)` calls are used to delegate to the parent constructor and methods.

# 🧠 Chellenge: Boss Fight!

Adding to `Player` and `Monster`, create a `Boss` subclass and modify the `Main.java` program to incorporate it and test it. Have fun in imagining the properties that the boss could have that would make it very different from the player and the monsters (and also hard to defeat!).

````python
%%writefile Boss.java
// Creating a Boss class monster that has additional features
public class Boss extends Entity {
    private boolean isUndead;
    private boolean isBoss;

    public Boss(String name, int maxHealth, boolean isUndead, boolean isBoss) {
        super(name, maxHealth);
        this.isUndead = isUndead;
        this.isBoss = isBoss;
    }

    @Override
    public void takeDamage(int amount) {
        if (isBoss) {
          // Boss monster reduces damage by 10
          amount -= 10;
          System.out.println(name + " is a boss and intimidates!");
        }
        if (isUndead) {
            amount /= 2;
            System.out.println(name + " is undead and resists damage!");
        }
        super.takeDamage(amount);
    }

    @Override
    public void heal(int amount) {
        if (isUndead) {
            System.out.println(name + " is undead and is hurt by healing!");
            takeDamage(amount); // Reuse takeDamage to simulate damage from healing
        } else {
            super.heal(amount);
        }
    }

    @Override
    public String toString() {
        return "Monster{name='" + name + "', undead=" + isUndead + ", isBoss=" + isBoss + ", health=" + health + "/" + maxHealth + "}";
    }
}
````
