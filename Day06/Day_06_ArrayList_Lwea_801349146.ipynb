# Example Java code using `ArrayList`

This example demonstrates how to use the `ArrayList` class in Java to manage a dynamic list of elements. It covers the basic operations you can perform on an `ArrayList` such as adding, accessing, modifying, and removing items.

## 📄 File: `ArrayListExample.java`

### 🔧 What It Does

This Java program:

1. Creates an `ArrayList` of Strings.
2. Adds fruit names to the list.
3. Accesses elements using index.
4. Modifies elements.
5. Removes items (by value and index).
6. Sorts the list.
7. Iterates over it using a loop.

## ✅ Concepts Covered

### 1. **Creating an ArrayList**

```java
ArrayList<String> fruits = new ArrayList<>();
```

- Uses `java.util.ArrayList`.
- The type inside `< >` defines what kind of data the list will store (`String` in this case).

---

### 2. **Adding Elements**

```java
fruits.add("Apple");
fruits.add("Banana");
```

- Adds elements to the end of the list.

### 3. **Accessing Elements**

```java
String first = fruits.get(0);
```

- Uses `.get(index)` to retrieve elements.

### 4. **Modifying Elements**

```java
fruits.set(1, "Blueberry");
```

- `.set(index, value)` updates an element at a specific index.

### 5. **Removing Elements**

```java
fruits.remove("Mango");    // By value
fruits.remove(0);          // By index
```

- Removes elements either by value or by position.

### 6. **Sorting**

```java
Collections.sort(fruits);
```

- Sorts the list alphabetically using `java.util.Collections`.

### 7. **Iterating Over Elements**

```java
for (String fruit : fruits) {
    System.out.println(fruit);
}
```

- Loops through each element using an enhanced `for` loop.

## 📌 Key Takeaways

- `ArrayList` is a **resizable array**.
- It's part of the **Java Collections Framework**.
- It allows **dynamic addition and removal** of items.
- Very useful when working with **variable-sized data collections**.

# The `ArrayListExample.java` program

Here is the code for the `ArrayListExample.java` program:

````python
%%writefile ArrayListExample.java
import java.util.ArrayList;
import java.util.Collections;

public class ArrayListExample {
    public static void main(String[] args) {

        // 1. Create an ArrayList of Strings
        ArrayList<String> fruits = new ArrayList<>();

        // 2. Add elements to the ArrayList
        fruits.add("Apple");
        fruits.add("Banana");
        fruits.add("Mango");
        fruits.add("Orange");

        // 3. Print the entire ArrayList
        System.out.println("Fruits list: " + fruits);

        // 4. Access elements by index
        System.out.println("First fruit: " + fruits.get(0)); // Apple

        // 5. Modify an element
        fruits.set(1, "Blueberry"); // Replaces Banana with Blueberry
        System.out.println("After modification: " + fruits);

        // 6. Remove an element by value
        fruits.remove("Mango");

        // 7. Remove an element by index
        fruits.remove(0); // Removes Apple

        System.out.println("After removals: " + fruits);

        // 8. Check if the list contains an element
        boolean hasOrange = fruits.contains("Orange");
        System.out.println("Contains Orange? " + hasOrange);

        // 9. Get index of an element
        int index = fruits.indexOf("Blueberry");
        System.out.println("Index of Blueberry: " + index);

        // 10. Loop using traditional for loop
        System.out.println("Loop using index:");
        for (int i = 0; i < fruits.size(); i++) {
            System.out.println("- " + fruits.get(i));
        }

        // 11. Loop using for-each
        System.out.println("Loop using for-each:");
        for (String fruit : fruits) {
            System.out.println("- " + fruit);
        }

        // 12. Check the size of the ArrayList
        System.out.println("List size: " + fruits.size());

        // 13. Check if the list is empty
        System.out.println("Is the list empty? " + fruits.isEmpty());

        // 14. Sort the ArrayList alphabetically
        Collections.sort(fruits);
        System.out.println("Sorted list: " + fruits);

        // 15. Clear all elements
        fruits.clear();
        System.out.println("After clearing: " + fruits);
        System.out.println("Is now empty? " + fruits.isEmpty());
    }
}
````

# 🧠 Challenges: Fruit Basket!

1. Add more elements like `"Pineapple"` or `"Watermelon"`.
2. Try removing a non-existent element (e.g., `fruits.remove("Grapes")`) and observe the result.
3. Try creating strategies to prevent elements that do already exist from being added.
4. Try creating strategies to prevent elements that do not exist from being removed.
5. Add a method to print the list in reverse order using a loop.

````python
%%writefile FruitBasket.java
import java.util.ArrayList;
import java.util.Collections;

public class FruitBasket {
    // Method for reverse list
    public static void reverse(ArrayList<String> fruits) {
        System.out.println("List in reverse:");
            for (int x = fruits.size() - 1; x >= 0; x--) {
              System.out.println("- " + fruits.get(x));
          }
    }

    // Remove repeats
    public static void removeDup(ArrayList<String> fruits) {
      for (int i = 0; i < fruits.size(); i++) {
            for (int j = i + 1; j < fruits.size(); j++) {
              if(fruits.get(i).equals(fruits.get(j))) {
                fruits.remove(j);
                j--;
              }
            }
          }
    }

    // Prevents false removals
    public static void newRemove(ArrayList<String> fruits, String newFruit) {
      int found = 0;
      for (int a = 0; a < fruits.size(); a++) {
        if (fruits.get(a).equals(newFruit)) {
          fruits.remove(newFruit);
          found = 1;
        }
      }
      if (found == 0) {
          System.out.println("element does not exist");
        }
    }

  public static void main(String[] args) {

        // 1. Create an ArrayList of Strings
        ArrayList<String> fruits = new ArrayList<>();

        // Add more elements to array list
        fruits.add("Apple");
        fruits.add("Banana");
        fruits.add("Mango");
        fruits.add("Orange");
        // add example elements
        fruits.add("Pineapple");
        fruits.add("Watermelon");

        // 3. Print the entire ArrayList
        System.out.println("Fruits list: " + fruits);

        // Remove a non-existant element (Nothing happens/removes nothing)
        fruits.remove("Grapes");
        System.out.println("Fruits list removing non-existant grapes: " + fruits);

        // Prevents duplicate elements from being added
        fruits.add("Apple");
        fruits.add("Banana");
        removeDup(fruits);
        System.out.println("Fruits list removing duplicates: " + fruits);

        // Prevents non-existant elements from being removed
        newRemove(fruits, "Grapes");

        //Reverse method in use
        reverse(fruits);
    }
}
````
