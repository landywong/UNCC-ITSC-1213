## Challenge 8. Encapsulation

## Challenge 10. Inheritance: super(...) call

`Animal` has no no-argument constructor: every `Animal` must be created with a name. The `Cat` subclass takes a name in its own constructor and is supposed to pass it up. Expected output is `Hello, I am Whiskers`.

Find the single compile-time bug. Edit the writefile cell
below, re-run it to overwrite the file on disk, then run
the compile-and-run cell that follows.

```python
%%writefile InheritanceFix.java
public class InheritanceFix {
    public static void main(String[] args) {
        Cat c = new Cat("Whiskers");
        System.out.println(c.greet());
    }
}

class Animal {
    protected String name;
    public Animal(String name) {
        this.name = name;
    }
}

class Cat extends Animal {
    public Cat(String name) {
        // expected: super(name);
        super(name);
    }
    public String greet() {
        return "Hello, I am " + name;
    }
}
```

```python
!javac InheritanceFix.java
!java InheritanceFix
```
