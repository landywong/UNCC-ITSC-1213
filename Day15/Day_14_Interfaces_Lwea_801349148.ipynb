# Interfaces in Java

Java interfaces allow you to define a contract that classes can implement. Interfaces are essential for organizing code, enabling polymorphism, and enforcing consistency.

---

## What Is an Interface?

- An **interface** is a list of method signatures (without implementations).
- It specifies *what* a class should do, but not *how*.
- Interfaces support **multiple implementations** and **polymorphism**.

### Example:

```java
public interface Playable {
    void play();
}
```

---

## Why Use Interfaces?

- Promotes **code reuse** and **modularity**
- Helps avoid repetitive `instanceof` checks and casting
- Makes your code more **extensible**: new types can plug into old code

### Before Using Interfaces:

```java
if (media instanceof MusicTrack) {
    ((MusicTrack) media).playMusic();
}
```

### After Using Interfaces:

```java
public void playMedia(Playable media) {
    media.play();
}
```

---

## Implementing an Interface

A class that implements an interface must provide implementations for all its methods.

```java
public class MusicTrack implements Playable {
    @Override
    public void play() {
        System.out.println("Playing music...");
    }
}
```

Use `@Override` to help catch typos or signature mismatches.

---

## Polymorphism with Interfaces

Once classes implement the same interface, they can be treated interchangeably.

```java
public static void playMedia(Playable media) {
    media.play();  // works with MusicTrack, PodcastEpisode, etc.
}
```

This is **polymorphism** in action: one method call, multiple behaviors depending on the object type.

---

## Design Recipe and Interfaces

Java interfaces align well with the Design Recipe used in software engineering:

- **Step 1**: Define your data (classes)
- **Step 2**: Specify method contracts (interfaces)
- **Step 3**: Compiler ensures you implement the contract
- **Step 4**: Reflect and expand the interface if needed

---

## Best Practices

- ✅ Use small, focused interfaces (Single Responsibility Principle)
- ✅ Use `@Override` for clarity and safety
- ✅ Document interface methods with `@param`, `@return`, and `@throws`
- ❌ Avoid large “god interfaces” with too many unrelated methods
- ❌ Don’t forget to implement all required methods

---

## Summary

Interfaces:

- Promote abstraction and separation of concerns
- Enable polymorphism
- Are enforced at compile time
- Help write flexible, testable, and extensible code



# Deomstration of the Usage of Interfaces in Java

## Write the `InterfaceDemo.java` script

````python
%%writefile InterfaceDemo.java
// InterfaceDemo.java

// Define an interface named Playable.
// This interface serves as a contract: any class that implements Playable
// must provide a method called play().
interface Playable {
    void play(); // This is a method signature, no body allowed
}

// First implementation of the Playable interface
class MusicTrack implements Playable {
    private String title;

    public MusicTrack(String title) {
        this.title = title;
    }

    // @Override ensures we are matching the interface exactly
    @Override
    public void play() {
        System.out.println("🎵 Playing music: " + title);
    }
}

// Second implementation of the Playable interface
class PodcastEpisode implements Playable {
    private String episodeTitle;

    public PodcastEpisode(String episodeTitle) {
        this.episodeTitle = episodeTitle;
    }

    @Override
    public void play() {
        System.out.println("🎙️ Streaming podcast episode: " + episodeTitle);
    }
}

// A third implementation, just for fun
class Audiobook implements Playable {
    private String bookTitle;

    public Audiobook(String bookTitle) {
        this.bookTitle = bookTitle;
    }

    @Override
    public void play() {
        System.out.println("📚 Narrating audiobook: " + bookTitle);
    }
}

// A generic player method that accepts any Playable object.
// Thanks to polymorphism, this method doesn't care which class implements Playable.
// As long as the object implements the interface, it can be used here.
public class InterfaceDemo {

    public static void playMedia(Playable item) {
        item.play(); // Calls the version of play() defined by the object's class
    }

    public static void main(String[] args) {
        // Create instances of each class
        MusicTrack track = new MusicTrack("Beethoven - Symphony No. 5");
        PodcastEpisode episode = new PodcastEpisode("Science Friday: Climate Tech");
        Audiobook book = new Audiobook("1984 by George Orwell");

        // Use the same method to play each item
        playMedia(track);
        playMedia(episode);
        playMedia(book);

        // Demonstrates how interfaces enable flexible, extensible code.
    }
}
````

````python
!javac InterfaceDemo.java
````

## Execute it

````python
!java InterfaceDemo
````

# Exercise 1: Fix the Bug in the Battle Simulation

You were given a basic RPG-style program using interfaces like `Attacker` and classes like `Cleric`, `Monster`, and `Player`. The program is below. However, a subtle exception-related bug was introduced, and you are challenged to fix it.

## Learning Goals

- Recognize the risk of null references.
- Use exceptions (e.g., `IllegalArgumentException`) to guard against invalid method parameters.
- Understand how to combine interface-based logic with defensive programming.

## Instructions

1. Identify the exception that occurs when running this code.
2.	Fix the `attack()` method in the `Player` class so that:
  - It throws an `IllegalArgumentException` if target is null.
	- It does not proceed with the attack in that case.
3.	Use a try-catch block in `main()` to handle this exception gracefully and print a helpful message.

````python
%%writefile Attacker.java
public interface Attacker {
    void attack(Character target);
}
````

````python
%%writefile Character.java
public class Character {
    private String name;
    private int health;

    public Character(String name, int health) {
        this.name = name;
        this.health = health;
    }

    public void takeDamage(int damage) {
        health -= damage;
    }

    public String getName() {
        return name;
    }

    public int getHealth() {
        return health;
    }
}
````

````python
%%writefile Player.java
public class Player extends Character implements Attacker {
    public Player(String name, int health) {
        super(name, health);
    }

    @Override
    public void attack(Character target) throws IllegalArgumentException {
        if (target == null) {
            throw new IllegalArgumentException("Target cannot be null!");
        }
        int damage = 10;
        System.out.println(getName() + " attacks " + target.getName() + " for " + damage + " damage!");
        target.takeDamage(damage);
    }
}
````

````python
%%writefile Main.java
public class Main {
    public static void main(String[] args) {
        Player hero = new Player("Artemis", 100);
        Character enemy = null;

        // Bug: We're not checking if 'enemy' is null!
        try {
            hero.attack(enemy);
        } catch (IllegalArgumentException e) {
            System.out.println("Error: " + e.getMessage());
        }
    }
}
````
