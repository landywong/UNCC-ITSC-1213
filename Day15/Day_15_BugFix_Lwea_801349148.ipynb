## Challenge 1. Interface declarations have no method bodies

**Topic: interfaces.** Declare a tiny interface and one class that implements it. Expected output:

```
Soft glow.
```

Find the single bug. (Hint: methods declared inside an interface must end with a semicolon -- they have no body.)

````python
%%writefile InterfaceBodyFix.java
public class InterfaceBodyFix {
    public static void main(String[] args) {
        Lightable lamp = new Lamp();
        lamp.lightUp();
    }
}

interface Lightable {
    // BUG: an interface method has no body. Remove the braces and
    //      end the declaration with a semicolon.
    void lightUp();
}

class Lamp implements Lightable {
    @Override
    public void lightUp() {
        System.out.println("Soft glow.");
    }
}
````

## Challenge 2. A class that implements an interface must supply every method

**Topic: interfaces.** `Greeter` declares two methods. The implementing class only writes one of them. Expected output:

```
Hello!
Goodbye!
```

Find the single bug.

````python
%%writefile MissingImplFix.java
public class MissingImplFix {
    public static void main(String[] args) {
        Greeter g = new EnglishGreeter();
        g.hello();
        g.goodbye();
    }
}

interface Greeter {
    void hello();
    void goodbye();
}

class EnglishGreeter implements Greeter {
    // BUG: this concrete class declares it implements Greeter
    //      but does not supply a body for goodbye(). The compiler
    //      will refuse to compile the class.
    @Override
    public void hello() {
        System.out.println("Hello!");
    }
    @Override
    public void goodbye() {
        System.out.println("Bye!");
    }
}
````

## Challenge 3. A single class may implement TWO interfaces

**Topic: interfaces.** `Bard` should implement BOTH `Singable` AND `Bonkable`. Expected output:

```
Tra-la-la!
Bonk!
```

Find the single bug. (Hint: a class can extend at most one class but may implement any number of interfaces. The two interfaces are listed comma-separated after `implements`.)

````python
%%writefile TwoInterfacesFix.java
public class TwoInterfacesFix {
    public static void main(String[] args) {
        Bard b = new Bard();
        b.sing();
        b.bonk();
    }
}

interface Singable { void sing(); }
interface Bonkable { void bonk(); }

// BUG: the header uses two `implements` clauses, which Java does
//      not allow. Use ONE `implements` followed by the two
//      interfaces separated by a comma.
class Bard implements Singable, Bonkable {
    @Override
    public void sing() { System.out.println("Tra-la-la!"); }
    @Override
    public void bonk() { System.out.println("Bonk!"); }
}
````

````python
!javac TwoInterfacesFix.java
!java TwoInterfacesFix
````

## Challenge 4. `try`/`catch`/`finally` ordering

**Topic: exceptions.** Catch the right exception type and let `finally` run. Expected output:

```
Caught: division by zero
Done.
```

Find the single bug. (Hint: dividing by zero throws `ArithmeticException`, not `NullPointerException`.)

````python
%%writefile CatchOrderFix.java
public class CatchOrderFix {
    public static void main(String[] args) {
        try {
            int x = 10 / 0;
            System.out.println(x);
        }
        // BUG: the wrong exception type is caught. 10 / 0 throws
        //      ArithmeticException, NOT NullPointerException, so
        //      the catch never matches and the program crashes.
        catch (ArithmeticException e) {
            System.out.println("Caught: " + e.getMessage());
        }
        finally {
            System.out.println("Done.");
        }
    }
}
````

````python
!javac CatchOrderFix.java
!java CatchOrderFix
````

## Challenge 5. Checked exceptions must be caught or declared

**Topic: exceptions.** `risky()` throws a CHECKED exception (`IOException`). Java refuses to compile any caller that ignores it. Expected output:

```
Caught: file gone
```

Find the single bug. (Hint: wrap the call in a try/catch.)

````python
%%writefile CheckedExceptionFix.java
import java.io.IOException;

public class CheckedExceptionFix {

    public static void risky() throws IOException {
        throw new IOException("file gone");
    }

    public static void main(String[] args) throws IOException {
        // BUG: risky() throws a CHECKED exception. The compiler
        //      will not let main() ignore it. Wrap the call in a
        //      try/catch (or declare main() with `throws IOException`).
        risky();
    }
}
````

## Challenge 6. `throw` versus `throws`

**Topic: exceptions.** Reject a negative deposit by throwing `IllegalArgumentException` (an UNCHECKED exception). Expected output:

```
Caught: amount must be non-negative
```

Find the single bug. (Hint: the keyword that *raises* an exception is `throw` -- no `s`. The `throws` clause goes on the method signature and is for *checked* exceptions only.)

````python
%%writefile ThrowVsThrowsFix.java
public class ThrowVsThrowsFix {

    public static void deposit(double amount) {
        if (amount < 0) {
            // BUG: `throws` is the keyword for the method signature.
            //      Inside a method body you must use `throw` (no `s`).
            throw new IllegalArgumentException("amount must be non-negative");
        }
        System.out.println("Deposited " + amount);
    }

    public static void main(String[] args) {
        try {
            deposit(-1.0);
        } catch (IllegalArgumentException e) {
            System.out.println("Caught: " + e.getMessage());
        }
    }
}
````

````python
!javac ThrowVsThrowsFix.java
!java ThrowVsThrowsFix
````

## Challenge 7. Abstract classes cannot be instantiated

**Topic: abstract classes.** `Shape` is abstract; `Square` is concrete. The driver tries to create one of each. Expected output:

```
16.0
```

Find the single bug.

````python
%%writefile AbstractInstantiateFix.java
public class AbstractInstantiateFix {
    public static void main(String[] args) {
        // BUG: Shape is abstract -- it cannot be instantiated with
        //      `new`. Drop this line entirely; only the Square is
        //      needed for the expected output.

        Square sq = new Square(4.0);
        System.out.println(sq.area());
    }
}

abstract class Shape {
    public abstract double area();
}

class Square extends Shape {
    private double side;
    public Square(double side) { this.side = side; }
    @Override
    public double area() { return this.side * this.side; }
}
````

````python
!javac AbstractInstantiateFix.java
!java AbstractInstantiateFix
````

## Challenge 8. Abstract methods declare a header, not a body

**Topic: abstract classes.** Declare an abstract method correctly. Expected output:

Find the single bug.

````python
%%writefile AbstractMethodFix.java
public class AbstractMethodFix {
    public static void main(String[] args) {
        Animal a = new Dog();
        a.speak();
    }
}

abstract class Animal {
    // BUG: an abstract method has NO BODY. The signature must end
    //      with a semicolon. Remove the empty braces.
    public abstract void speak();
}

class Dog extends Animal {
    @Override
    public void speak() {
        System.out.println("Wuff wuff!");
    }
}
````

## Challenge 9. A concrete subclass must implement every abstract method

**Topic: abstract classes.** `Vehicle` declares two abstract methods. `Car` only implements one of them. Expected output:

```
Vroom!
Stopping.
```

Find the single bug.

````python
%%writefile ConcreteSubclassFix.java
public class ConcreteSubclassFix {
    public static void main(String[] args) {
        Vehicle v = new Car();
        v.start();
        v.stop();
    }
}

abstract class Vehicle {
    public abstract void start();
    public abstract void stop();
}

// BUG: Car claims to be a CONCRETE subclass of an abstract parent
//      with TWO abstract methods, but it only implements start().
//      A concrete subclass must implement EVERY inherited abstract
//      method. Add the missing override of stop().
class Car extends Vehicle {
    @Override
    public void start() {
        System.out.println("Vroom!");
    }
    @Override
    public void stop() {
        System.out.println("Stopping.");
    }
}
````
