# Exceptions and Errors in Java

## What is an Exception?

An exception is an event that disrupts the normal flow of a program. In Java, exceptions are objects that describe an error or unusual condition.

Java has two main types of exceptions:
	1.	Checked Exceptions â€“ Must be declared or handled.
	2.	Unchecked Exceptions â€“ Donâ€™t need to be declared or handled (e.g., `NullPointerException`).

## How are Exceptions different than Errors?

### The Throwable Hierarchy

```
             Throwable
             /       \
       Exception     Error
```

### Exceptions

These are conditions your program can catch and potentially recover from.

**Two types:**

- **Checked exceptions** (must be declared or caught)
  - e.g., IOException, SQLException
- **Unchecked exceptions** (donâ€™t need to be declared or caught)
  - Subclass of RuntimeException
  - e.g., NullPointerException, IllegalArgumentException

**Use case:**

- Missing file
- Invalid user input
- Network interruption

You can handle exceptions using try-catch, or propagate them using throws.

### Errors

These are serious problems that a program should not try to handle.

**Examples:**

- `OutOfMemoryError`
- `StackOverflowError`
- `InternalError`
- `NoClassDefFoundError`

**Use case:**

- JVM runs out of memory
- Infinite recursion causes stack overflow
- Class loading failure

You generally donâ€™t catch Errors, because they indicate fundamental issues with the runtime environment. Trying to recover from them is often pointless or dangerous.

## Suggested Java Exception Classes

Words like `IOException` are not Java keywords (like `public`, `class`, `if`, or `for`), but they are class names from the Java standard library, specifically from the `java.io` or `java.lang` packages.

### How do you know which exception class to use?

Java provides a toolbox of built-in exceptions, each representing a specific type of problem. You choose the tool (exception class) that best matches the type of error youâ€™re handling.

Hereâ€™s a table to help you decide (but, when in doubt, remember that you can start with a general one):

If none of the built-in ones fit, you can define your own. For example:

```java

    }

```

## Notes
- **Checked exceptions** must be declared in the method signature using `throws`, or handled with `try-catch`.
- **Unchecked exceptions** (subclasses of `RuntimeException`) donâ€™t need to be declared or caught, but can be.
- When creating custom exceptions, you usually extend `Exception` or `RuntimeException`.


## Summary


- **Use exceptions to handle expected issues** (like missing files, invalid input).
- **Donâ€™t catch errors** â€” they signal JVM-level failures that your program shouldnâ€™t try to fix.

# Exceptions Demonstration

The program below, `ExceptionDemo.java`, contains a class to demonstrate different types of exceptions in Java.

````python
%%writefile ExceptionDemo.java
import java.io.*;

// Main class to demonstrate different types of exceptions in Java
public class ExceptionDemo {

    // === CHECKED EXCEPTION EXAMPLE ===
    // This method simulates reading a file. If the file doesn't exist, it throws IOException.
    public static void readFile(String filename) throws IOException {
        // Simulate a checked exception by trying to open a non-existent file
        FileReader reader = new FileReader(filename); // May throw FileNotFoundException
        reader.close();
    }

    // === UNCHECKED EXCEPTION EXAMPLE ===
    // This method divides two integers. If the denominator is zero, an ArithmeticException is thrown.
    public static int divide(int a, int b) {
        return a / b; // Division by zero is an unchecked exception (ArithmeticException)
    }

    // === ERROR EXAMPLE ===
    // This method calls itself infinitely, causing a StackOverflowError.
    public static void recursiveCall() {
        recursiveCall(); // Not caught normally; demonstrates a JVM error
    }

    public static void main(String[] args) {
        System.out.println("=== Checked Exception Example ===");
        try {
            readFile("nonexistent.txt"); // This file doesn't exist
        } catch (IOException e) {
            System.out.println("Caught checked exception: " + e.getMessage());
        }

        System.out.println("\n=== Unchecked Exception Example ===");
        try {
            int result = divide(10, 0); // Triggers ArithmeticException
            System.out.println("Result: " + result);
        } catch (ArithmeticException e) {
            System.out.println("Caught unchecked exception: " + e.getMessage());
        }

        System.out.println("\n=== Error Example ===");
        try {
            recursiveCall(); // WARNING: this will crash the program!
        } catch (StackOverflowError e) {
            System.out.println("Caught error (rare): " + e.getMessage());
        }

        System.out.println("\nProgram ended gracefully.");
    }
}
````

# Examples

## Reading a File

````python
// Before (no exception handling)
public static void readFile(String filename) {
    FileReader reader = new FileReader(filename);  // Compile error: unhandled exception
    reader.close();
}
````

````python
// After (with throwns)
public static void readFile(String filename) throws IOException {
    FileReader reader = new FileReader(filename);
    reader.close();
}
````

````python
// After (with try-catch)
public static void readFile(String filename) {
    try {
        FileReader reader = new FileReader(filename);
        reader.close();
    } catch (IOException e) {
        System.out.println("Error reading file: " + e.getMessage());
    }
}
````

## Dividing Two Numbers

````python
// Before (no exception handling)
public static int divide(int a, int b) {
    return a / b;  // Risk: division by zero!
}
````

````python
// After (handle runtime exception)
public static int divide(int a, int b) {
    try {
        return a / b;
    } catch (ArithmeticException e) {
        System.out.println("Cannot divide by zero!");
        return 0; // default value
    }
}
````

## Custom Exception

````python
// Before (manual error handling)
public static void checkAge(int age) {
    if (age < 18) {
        System.out.println("Underage!");
    }
}
````

````python
//  After (using custom exception)
public static void checkAge(int age) throws UnderageException {
    if (age < 18) {
        throw new UnderageException("You must be 18 or older.");
    }
}
````

````python
// And the custom exception:
public class UnderageException extends Exception {
    public UnderageException(String message) {
        super(message);
    }
}
````

## Catching Multiple Exceptions

````python
public static void loadData(String filename) {
    try {
        FileReader reader = new FileReader(filename);
        int result = 10 / 0;
        reader.close();
    } catch (FileNotFoundException e) {
        System.out.println("File not found!");
    } catch (ArithmeticException e) {
        System.out.println("Math error: " + e.getMessage());
    } catch (IOException e) {
        System.out.println("IO error: " + e.getMessage());
    }
}
````
