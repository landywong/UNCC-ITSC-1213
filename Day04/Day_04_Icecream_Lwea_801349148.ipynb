### The String in Java

In Java, a String is a **sequence of characters**, like "Hello" or "Java is fun". It’s used to represent **textual data**.

Key Facts:

- Strings are **objects**, not primitive types.
- Strings are **immutable**: once created, they cannot be changed.
- You create Strings using **double quotes**, like this:

```java
String greeting = "Hello, world!";
```

How to create a String:

```java
String s1 = "Hello"; // most common way
String s2 = new String("Hello"); // less common, creates a new object
```

- `s1` uses the string pool (Java caches string literals).
- `s2` creates a new object, even if the text is the same

Here are some useful methods you can call on a String:

| **Method**                    | **What it Does**                      | **Example**                          |
| ----------------------------- | ------------------------------------- | ------------------------------------ |
| length()                      | Returns the number of characters      | "Hello".length() → 5                 |
| charAt(index)                 | Gets the character at a position      | "Java".charAt(0) → 'J'               |
| substring(start, end)         | Extracts a portion of the string      | "Hello".substring(0, 2) → "He"       |
| equals(other)                 | Compares content of strings           | "abc".equals("abc") → true           |
| equalsIgnoreCase(other)       | Case-insensitive comparison           | "Hello".equalsIgnoreCase("hello")    |
| toLowerCase() / toUpperCase() | Changes case                          | "Hi".toUpperCase() → "HI"            |
| contains(str)                 | Checks if a string includes another   | "Java".contains("va") → true         |
| replace(old, new)             | Replaces parts of the string          | "Java".replace("a", "o") → "Jovo"    |
| split(delimiter)              | Splits string into array by delimiter | "a,b,c".split(",") → ["a", "b", "c"] |
| trim()                        | Removes whitespace from both ends     | " hi ".trim() → "hi"                 |



String Comparison: `==` vs `.equals()`



```java
String a = "hello";
String b = "hello";
String c = new String("hello");

a == b       // true: both point to the same object in the string pool
a == c       // false: c is a new object
a.equals(c)  // true: compares the actual text
```

Immutability! Once you create a string, you cannot change it. Methods like `replace()` or `toUpperCase()` return a new string:

```java
String original = "hello";
String upper = original.toUpperCase();
System.out.println(original); // still "hello"
System.out.println(upper);    // "HELLO"
```

If you need to build or modify strings many times, use `StringBuilder`:

```java
StringBuilder sb = new StringBuilder();
sb.append("Hello");
sb.append(" World");
System.out.println(sb.toString()); // "Hello World"
```

`StringBuilder` is faster than doing str += "something" repeatedly.

### What is a Constructor in Java?

A constructor is a special method that is called automatically when you create an object from a class. Its main job is to initialize the object — that is, set up its initial values.

| **Feature**        | **Description**                                              |
| ------------------ | ------------------------------------------------------------ |
| Same name as class | A constructor must have the **exact same name** as the class |
| No return type     | It has **no return type**, not even void                     |
| Called with new    | You **invoke it when creating an object** using the new keyword |
| Can be overloaded  | A class can have **multiple constructors** with different parameters |



Example:



```java
public class Dog {
    String name;
    int age;

    // Constructor
    public Dog(String name, int age) {
        this.name = name;
        this.age = age;
    }

    public void bark() {
        System.out.println(name + " says woof!");
    }
}
```



Using the constructor:



```java
Dog myDog = new Dog("Buddy", 3);  // creates and initializes
myDog.bark();                     // Buddy says woof!
```



Note that, if you don’t write any constructor, Java creates one for you with no parameters:



```java
public class Cat {
    String color;
}

// Java provides: public Cat() {}
```



You can have multiple constructors with different numbers or types of parameters:



```java
public class Icecream {
    String flavor;
    int scoops;

    // Constructor 1
    public Icecream() {
        flavor = "vanilla";
        scoops = 1;
    }

    // Constructor 2
    public Icecream(String flavor, int scoops) {
        this.flavor = flavor;
        this.scoops = scoops;
    }
}
```

Now both are valid:

```java
Icecream a = new Icecream();                // uses default
Icecream b = new Icecream("chocolate", 3);  // uses second
```

# The `Icecream.java` program

Here is the code for `Icecream.java`:

````python
%%writefile Icecream.java
public class Icecream {
    // FIELDS
    private String flavor;
    private int scoopCount;
    private boolean cone;

    // Constructor
    public Icecream() {
    }

    public Icecream(String flav, int scpCount, boolean cone) {
        this.flavor = flav;
        this.scoopCount = scpCount;
        this.cone = cone;
    }

    // GETTER FOR FLAVOR
    public String getFlavor() {
        return this.flavor;
    }

    // GETTER FOR SCOOPCOUNT
    public int getScoopCount() {
        return this.scoopCount;
    }

    // GETTER FOR CONE
    public boolean isCone() {
        return this.cone;
    }

    // SETTER FOR FLAVOR
    public void setFlavor(String flav) {
        this.flavor = flav;
    }

    // SETTER FOR SCOOPCOUNT
    public void setScoopCount(int cnt) {
        this.scoopCount = cnt;
    }

    // SETTER FOR CONE
    public void setCone(boolean cone) {
        this.cone = cone;
    }

    // METHOD FOR ADDING A SCOOP
    public int addScoop() {
        this.scoopCount++;
        return this.scoopCount;
    }

    // METHOD TO DISPLAY THE ICECREAM DETAILS
    public void details() {
        System.out.println("This is a " + this.flavor + " icecream with "
                + this.scoopCount + " scoops");
        if (cone) {
            System.out.println(" and is on a cone.");
        } else {
            System.out.println(" and is in a cup not a cone.");
        }
    }

}// end class
````

# The `Example.java` program

Here is the code for the `Example.java` program:

````python
%%writefile Example.java
public class Example {

  public static void main(String[] args) {

    // first Icecream object using default constructor
    Icecream myIcecream = new Icecream();

    System.out.println("CREATE ICECREAM 1: CHOCO");

    myIcecream.setScoopCount(10);
    myIcecream.setFlavor("chocolate");
    myIcecream.setCone(true);
    System.out.println(myIcecream.getFlavor());
    System.out.println(myIcecream.getScoopCount());
    System.out.println(myIcecream.isCone());

    System.out.println("CREATE ICECREAM 2: FRESA");

    // second Icecream object using second constructor
    Icecream yourIcecream = new Icecream("strawberry", 53, false);
    System.out.println(yourIcecream.getFlavor());
    System.out.println(yourIcecream.getScoopCount());
    System.out.println(yourIcecream.isCone());

    // store Icecream objects in an array
    Icecream[] orders = new Icecream[2];
    orders[0] = myIcecream;
    orders[1] = yourIcecream;

    System.out.println("LOOP 1");

    // find the total number of scoops
    int count = 0;
    for (Icecream theIcream : orders) {
      count = count + theIcream.getScoopCount();
    }
    System.out.println("For loop type 1: Total scoops sold: " + count);

    System.out.println("LOOP 2");

    // using a regular for loop
    count = 0; // reset count
    for (int i = 0; i < orders.length; i++){
      Icecream theIcecream = orders[i]; // intermediate variable to reference the object
      count += theIcecream.getScoopCount(); // using compound operator +=
    }
    System.out.println("For loop type 2: Total scoops sold: " + count);

    System.out.println("WHILE LOOP");

    // using a while loop
    count = 0; //reset count
    int j = 0;
    while (j < orders.length) {
      count += orders[j].getScoopCount(); //accessing the object from the array without using an intermediate variable.
      j++;
    }
    System.out.println("While loop: Total scoops sold: " + count);

    System.out.println("CREATE ICECREAM 3: VANILLA");

    Icecream icecreamObj = new Icecream("vanilla", 2, true);
    icecreamObj.details();

    System.out.println("...MODIFIED THE CONE...");

    icecreamObj.setCone(false);
    // display object new state
    icecreamObj.details();

    // call for the addScoop method
    icecreamObj.addScoop();

    System.out.println("...ADDED A SCOOP...");

    // display object new state
    icecreamObj.details();

  }// end main
}// end class
````
