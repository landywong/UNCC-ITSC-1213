Find the single bug. Edit the writefile cell below, re-run it to overwrite the file on disk, then run the compile-and-run cell that follows.

````python
%%writefile PrivateFieldFix.java
public class PrivateFieldFix {
    public static void main(String[] args) {
        Movie m = new Movie("Inception");
        // BUG: title is private; outside code cannot read it directly.
        // change .title to .getTitle()
        System.out.println(m.getTitle());
    }
}

class Movie {
    private String title;
    public Movie(String title) { this.title = title; }
    public String getTitle() { return this.title; }
}
````

## Challenge 2. Reference assignment is aliasing, not copying

**Module 3 (classes and objects).** The program is supposed to leave `original.n` at 5 and set `copy.n` to 99. Expected output:

```
original.n = 5
copy.n     = 99
```

Find the single bug.

````python
%%writefile AliasingFix.java
public class AliasingFix {
    public static void main(String[] args) {
        Box original = new Box(5);
        // BUG: this is aliasing -- `copy` and `original` point at the SAME Box.
        // changed copy to new Box rather that same object
        Box copy = new Box(5);
        copy.n = 99;
        System.out.println("original.n = " + original.n);
        System.out.println("copy.n     = " + copy.n);
    }
}

class Box {
    int n;
    public Box(int n) { this.n = n; }
}
````

````python
!javac AliasingFix.java
!java AliasingFix
````

## Challenge 5. `this` inside a `static` method

**Module 4 (beyond the basics).** Increment a class-wide counter three times and print the result. Expected output: `3`.

Find the single bug.

````python
%%writefile StaticThisFix.java
public class StaticThisFix {
    private static int count = 0;

    // BUG: `this` cannot be used inside a static method --
    // there is no current object for it to refer to.
    // removed this from count
    public static int increment() {
        count++;
        return count;
    }

    public static void main(String[] args) {
        increment();
        increment();
        System.out.println(increment());
    }
}
````

**Module 5 (Strings and ArrayList).** Convert `"hello"` to upper case and print the result. Expected output: `HELLO`.

The program compiles, but the output is wrong. Find the single bug.

## Challenge 8. `ArrayList.set` versus `ArrayList.add`

**Module 5 (Strings and ArrayList).** Build an empty `ArrayList<String>`, put two strings into it, and print them. Expected output:

```
first second
```

Find the single bug. (Hint: this one is a runtime error, not a compile error -- read the exception's stack trace carefully.)

````python
%%writefile ArrayListFix.java
import java.util.ArrayList;

public class ArrayListFix {
    public static void main(String[] args) {
        ArrayList<String> list = new ArrayList<>();
        // BUG: set(i, x) replaces an existing slot. The list is empty,
        //      so index 0 does not yet exist -- this throws.
        // change set to add
        list.add(0, "first");
        list.add("second");
        System.out.println(list.get(0) + " " + list.get(1));
    }
}
````

````python
!javac ArrayListFix.java
!java ArrayListFix
````

## Challenge 9. A subclass constructor missing `super(...)`

**Module 6 (inheritance).** Build a `Cat` named `"Whiskers"` and have it greet. Expected output: `Hello, I am Whiskers`.

Find the single bug. (Hint: read the compiler error carefully -- it points at the constructor.)

````python
%%writefile SuperCallFix.java
public class SuperCallFix {
    public static void main(String[] args) {
        Cat c = new Cat("Whiskers");
        System.out.println(c.greet());
    }
}

class Animal {
    protected String name;
    // Animal has NO no-argument constructor.
    public Animal(String name) { this.name = name; }
}

class Cat extends Animal {
    // BUG: the implicit super() call fails because Animal has
    //      no no-arg constructor. The subclass must call super(name)
    //      explicitly.
    // added super name
    public Cat(String name) {
      super(name);
    }
    public String greet() { return "Hello, I am " + name; }
}
````

````python
!javac SuperCallFix.java
!java SuperCallFix
````

## Challenge 10. `@Override` catches a typo

**Module 6 (inheritance).** Override `Dessert.describe()` in `Sundae`. Expected output: `A sundae.`

Find the single bug. (Hint: the `@Override` annotation is doing exactly the job it was added for -- read the compiler error.)

````python
%%writefile OverrideFix.java
public class OverrideFix {
    public static void main(String[] args) {
        Sundae s = new Sundae();
        s.describe();
    }
}

class Dessert {
    public void describe() {
        System.out.println("A dessert.");
    }
}

class Sundae extends Dessert {
    // BUG: typo in the method name. With @Override, the compiler
    //      catches that this method does NOT override anything.
    // edit typo in "describe"
    @Override
    public void describe() {
        System.out.println("A sundae.");
    }
}
````

````python
!javac OverrideFix.java
!java OverrideFix
````
