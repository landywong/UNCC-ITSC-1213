# READ-ME

## The Person-Student Java Project

This Java project demonstrates the core principles of **object-oriented programming (OOP)** using inheritance. It consists of a superclass `Person` and a subclass `Student`, where the `Student` class extends the basic identity information of a person with attributes relevant to academic life. The project also includes a test class `TestStudent` that creates and manipulates student objects, demonstrating constructors, getters/setters, and list operations.

## Features

- Encapsulation of identity data in a `Person` class
- Inheritance: `Student` extends `Person`
- Custom constructor that initializes both inherited and new fields
- GPA, credit, and balance tracking for students
- Course enrollment system using `ArrayList<String>`
- Methods to add, drop, and list courses
- Boolean return for dropping a course (to indicate success/failure)
- Comprehensive test class to demonstrate functionality

# Part 1: `Person.java`

````python
%%writefile Person.java
public class Person {

    private String firstName;
    private String lastName;
    private int id;

    public Person(String firstName, String lastName, int id) {
        this.firstName = firstName;
        this.lastName = lastName;
        this.id = id;
    }

    /*
     * Get the value of firstName
     *
     * @return the value of firstName
     */

    public String getFirstName() {
        return firstName;
    }

    /*
     * Get the value of lastName
     *
     * @return the value of lastName
     */

    public String getLastName() {
        return lastName;
    }

    /*
     * Get the value of id
     *
     * @return the value of id
     */

    public int getId() {
        return id;
    }

    public void display() {
        System.out.println("Name: " + this.firstName + " " + this.lastName);
        System.out.println("ID: " + this.id);
    }
}
````

# Part 3: `Student.java`

````python
%%writefile Student.java
import java.util.ArrayList;

/*
 * The Student class extends the Person class and adds academic-related attributes and behaviors.
 * It includes fields for major, GPA, credits, balance, transfer status, and enrolled courses.
 */

public class Student extends Person {

    // Instance variables unique to Student
    private String major;
    private double gpa;
    private int credits;
    private double balance;
    private boolean transfer;
    private ArrayList<String> enrolledCourses;

    /*
     * Constructs a Student object by initializing fields from Person and Student.
     *
     * @param firstName  the student's first name (passed to Person)
     * @param lastName   the student's last name (passed to Person)
     * @param id         the student ID (passed to Person)
     * @param major      the student's declared major
     * @param gpa        the student's GPA
     * @param credits    the number of credits earned
     */

    public Student(String firstName, String lastName, int id,
                   String major, double gpa, int credits) {
        super(firstName, lastName, id);  // Call Person constructor
        this.major = major;
        this.gpa = gpa;
        this.credits = credits;
        this.balance = 0.0;
        this.transfer = false;
        this.enrolledCourses = new ArrayList<>();
    }

    /* Returns the student's major. */

    public String getMajor() {
        return major;
    }

    /* Returns the student's GPA. */

    public double getGPA() {
        return gpa;
    }

    /* Returns the student's earned credits. */

    public int getCredits() {
        return credits;
    }

    /* Returns whether the student is a transfer student. */

    public boolean isTransfer() {
        return transfer;
    }

    /* Returns the student's current balance. */

    public double getBalance() {
        return balance;
    }

    /* Returns the list of enrolled course names. */

    public ArrayList<String> getEnrolledCourses() {
        return enrolledCourses;
    }

    /* Sets the student's major. */

    public void setMajor(String major) {
        this.major = major;
    }

    /* Sets the student's GPA. */

    public void setGPA(double gpa) {
        this.gpa = gpa;
    }

    /* Sets the number of credits the student has earned. */

    public void setCredits(int credits) {
        this.credits = credits;
    }

    /* Sets whether the student is a transfer student. */

    public void setTransfer(boolean transfer) {
        this.transfer = transfer;
    }

    /* Sets the student's balance. */

    public void setBalance(double balance) {
        this.balance = balance;
    }

    /*
     * Adds a course to the student's enrolledCourses list.
     *
     * @param course the course name to add
     */

    public void addCourse(String course) {
        enrolledCourses.add(course);
    }

    /*
     * Removes a course from the student's enrolledCourses list.
     *
     * @param course the course name to remove
     * @return true if the course was removed, false if it wasn't found
     */

    public boolean dropCourse(String course) {
        return enrolledCourses.remove(course);
    }
}
````

# Part 4: `TestStudent.java`

````python
%%writefile TestStudent.java
import java.util.ArrayList;

public class TestStudent {
    public static void main(String[] args) {
        System.out.println("***** Testing Student *****");

        Student s1 = new Student("Xavier", "Cato", 9000, "CS", 3.5, 120);
        s1.setTransfer(true);
        s1.setBalance(100);
        s1.addCourse("Java");
        s1.addCourse("Calculus");
        s1.addCourse("Spanish");

        s1.display();

        System.out.println("credits: " + s1.getCredits());
        System.out.println("GPA: " + s1.getGPA());
        System.out.println("Major: " + s1.getMajor());
        System.out.println("Transfer: " + s1.isTransfer());

        s1.setCredits(0);
        s1.setGPA(0);
        s1.setMajor("Math");

        System.out.println("credits: " + s1.getCredits());
        System.out.println("GPA: " + s1.getGPA());
        System.out.println("Major: " + s1.getMajor());

        if (s1.dropCourse("Spanish")) {
            System.out.println("the class has been dropped");
        } else {
            System.out.println("the class does not exist");
        }

        ArrayList<String> enrolledCourses = s1.getEnrolledCourses();
        System.out.println("You are enrolled in the following courses:");
        for (String course : enrolledCourses) {
            System.out.println(course);
        }
    }
}
````

### 🔨 Methods to Implement

````python
%%writefile Professor.java
import java.util.ArrayList;
import java.util.Collections;

/*
 * The Professor class extends the Person class and adds academic-related attributes and behaviors.
 * It includes fields for department, officeNumber, tenured, coursesTaught.
 */

public class Professor extends Person {
    // Instance variables unique to Professor
    private String department;
    private String officeNumber;
    private boolean tenured;
    private ArrayList<String> coursesTaught;

/*
     * Constructs a Professor object by initializing fields from Person and Student.
     *
     * @param firstName       the professor's first name (passed to Person)
     * @param lastName        the professor's last name (passed to Person)
     * @param id              the professor ID (passed to Person)
     * @param department      the professor's department
     * @param officeNumber    the professor's officeNumber
     * @param tenured         if professor is tenured
     * @param coursesTaught   professor's list of courses taught
     */

    public Professor(String firstName, String lastName, int id, String department, String officeNumber, boolean tenured) {
        super(firstName, lastName, id);  // Call Person constructor
        this.department = department;
        this.officeNumber = officeNumber;
        this.tenured = tenured;
        this.coursesTaught = new ArrayList<>();
    }

    // Getter and setter for department
    public String getDepartment() {
        return department;
    }
    public void setDepartment(String department) {
        this.department = department;
    }

    // Getter and setter for office number
    public String getOfficeNumber() {
        return officeNumber;
    }
    public void setOfficeNumber(String officeNumber) {
        this.officeNumber = officeNumber;
    }

    // Getter and setter for tenured
    public boolean isTenured() {
      return tenured;
    }
    public void setTenured(boolean tenured) {
      this.tenured = tenured;
    }

    // Method for adding course to taught list
    public void addCourse(String course) {
        coursesTaught.add(course);
    }

    // Method for removing course from taught list
    public boolean dropCourse(String course) {
        return coursesTaught.remove(course);
    }

    // Method for returning courses taught, sorted alphabetically
    public ArrayList<String> getCoursesTaught() {
      Collections.sort(coursesTaught);
      return coursesTaught;
    }

    // Displays professor info
    @Override
    public void display() {
        super.display();
        System.out.println("Department: " + this.department);
        System.out.println("Office: " + this.officeNumber);
    }
}
````

````python
%%writefile TestProfessor.java
import java.util.ArrayList;

public class TestProfessor {
    public static void main(String[] args) {
        System.out.println("***** Testing Professor(s) *****");

        // Array list of professors
        ArrayList<Professor> profList = new ArrayList<>();

        Professor a = new Professor("John", "Smith", 1825, "CS", "E003", true);
        profList.add(a);
        Professor b = new Professor("Sarah", "Wright", 1032, "History", "C024", true);
        profList.add(b);
        Professor c = new Professor("Wanda", "Rome", 2041, "Math", "B125", false);
        profList.add(c);

        // Adding courses taught
        a.addCourse("Data Structures");
        a.addCourse("Intro to CS");

        b.addCourse("American History");
        b.addCourse("World History");

        c.addCourse("Calculus");
        c.addCourse("Statistics");

        //Test getter & setters
        System.out.println("Prof. Smith courses: " + a.getCoursesTaught());

        System.out.println("Prof. Rome tenured? " + c.isTenured());
        c.setTenured(true);
        System.out.println("Prof. Rome tenured v2? " + c.isTenured());

        // Display loop for professors
        System.out.println("\nList of Professors:");
        for (Professor x : profList) {
          x.display();
          System.out.println();
        }
    }
}
````
