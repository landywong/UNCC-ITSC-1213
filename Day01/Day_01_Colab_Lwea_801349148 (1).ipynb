# Running Java programs using Google Colabs

You can run `Java` code by compiling and executing it using `shell` commands within a Colab notebook.


## Here are a few tips to start using Google Colabs

1. Write `%%writefile MyJavaProgram.java` in line one to write your code into a file named `MyJavaProgram.java` (change the file name as needed).

```python
%%writefile MyJavaProgram.java
public class MyJavaProgram {
    public static void main(String[] args) {
        System.out.println("Hello from this Java file!");
    }
}
```

2. To run `shell` commands, start each command line with an exclammation mark `!`. For example, do `!javac MyJavaProgram.java` to compile `MyJavaProgram.java` and `ls` to list the contents of the working direcotry after compilarion.

```python
!javac MyJavaProgram.java
!ls
```

3. To run the program after compilation, execute `!java MyJavaProgram`

```python
!java MyJavaProgram
```
