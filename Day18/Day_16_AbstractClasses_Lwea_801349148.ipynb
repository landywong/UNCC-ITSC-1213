# The `Player.java` script

````python
%%writefile Player.java
/*
 * Purpose: Represents a player in the game with health, name, level and special ability.
 */

public class Player extends Character{

    private int level;
    private String specialAbility;

    public Player(String name, int health, int level, String specialAbility) {
        super(name, health);
        this.level = level;
        this.specialAbility = specialAbility;
    }

    public int getLevel() {
        return level;
    }

    public void setLevel(int level) {
        this.level = level;
    }

    public String getSpecialAbility() {
        return specialAbility;
    }

    public void setSpecialAbility(String specialAbility) {
        this.specialAbility = specialAbility;
    }

    public void display() {
        System.out.println("Player: " + getName() + ", Health: " + getHealth() + ", Level: " + getLevel() + ", Ability: " + specialAbility);
    }

    @Override
    public String toString() {
        return "Player{name='" + getName() + "', health=" + getHealth() +
               ", level=" + level + ", s.ability='" + specialAbility + "'}";
    }

    @Override
public boolean equals(Object other) {
    if (!(other instanceof Player)) return false;
    Player p = (Player) other;
    return this.getName().equalsIgnoreCase(p.getName());
}
}
````

# The `Monster.java` script

````python
%%writefile Monster.java
/*
 * Purpose: Represents a monster with health, name, type and damage
 */

public class Monster extends Character {

    private String type;
    private int damage;

    public Monster(String name, int health, String type, int damage) {
        super(name, health);
        this.type = type;
        this.damage = damage;
    }


    public String getType() {
        return type;
    }

    public void setType(String type) {
        this.type = type;
    }

    public int getDamage() {
        return damage;
    }

    public void setDamage(int damage) {
        this.damage = damage;
    }

    public void display() {
        System.out.println("Monster: " + getName() + ", Health: " + getHealth() + ", Type: " + type + ", Damage: " + damage);
    }

    @Override
    public String toString() {
        return "Monster{name='" + getName() + "', health=" + getHealth() +
               ", type='" + type + "', damage=" + damage + "}";
    }

    @Override
public boolean equals(Object other) {
    if (!(other instanceof Monster)) return false;
    Monster m = (Monster) other;
    return this.getName().equalsIgnoreCase(m.getName());
}
}
````

# The `Character.java` program

````python
%%writefile Character.java
/*
 * Purpose: Represents a character with health and name.
 */

 public abstract class Character {
  private String name;
  private int health;

  public Character(String name, int health) {
        this.name = name;
        this.health = health;
    }

    public String getName() {
        return name;
    }

    public void setName(String name) {
        this.name = name;
    }

    public int getHealth() {
        return health;
    }

    public void setHealth(int health) {
        this.health = health;
    }

    public abstract void display();

    @Override
    public String toString() {
        return "Character{name='" + name + "', health=" + health + "}";
    }

    @Override
    public boolean equals(Object other) {
        if (!(other instanceof Character)) return false;
        Character c = (Character) other;
        return this.name.equalsIgnoreCase(c.name);
    }
 }

// Add your code in here
````
