````python
%%writefile Character.java
public abstract class Character {
    private String name;
    private int health;

    public Character(String name, int health) {
        this.name = name;
        this.health = health;
    }

    public String getName() {
        return name;
    }

    public int getHealth() {
        return health;
    }

    public void setHealth(int health) {
        // prevent negative health
        this.health = Math.max(0, health);
    }

    // Abstract method that each subclass must override
    public abstract void display();
}
````

````python
%%writefile Player.java
public class Player extends Character implements Attacker {

    private int level;
    private String classType;

    public Player(String name, int health, int level, String classType) {
        super(name, health);
        this.level = level;
        this.classType = classType;
    }

    public int getLevel() {
        return level;
    }

    public String getClassType() {
        return classType;
    }

    @Override
    public void display() {
        System.out.println("=== PLAYER ===");
        System.out.println("Name: " + getName());
        System.out.println("Health: " + getHealth());
        System.out.println("Level: " + level);
        System.out.println("Class Type: " + classType);
    }

    @Override
    public void attack(Character target) {
        int damage = 10;
        System.out.println(getName() + " attacks " + target.getName() + " for " + damage + " damage!");
        target.setHealth(target.getHealth() - damage);
    }
}
````

````python
%%writefile Monster.java
public class Monster extends Character implements Attacker {

    private String type;
    private boolean isBoss;

    public Monster(String name, int health, String type, boolean isBoss) {
        super(name, health);
        this.type = type;
        this.isBoss = isBoss;
    }

    public String getType() {
        return type;
    }

    public boolean isBoss() {
        return isBoss;
    }

    @Override
    public void display() {
        System.out.println("=== MONSTER ===");
        System.out.println("Name: " + getName());
        System.out.println("Health: " + getHealth());
        System.out.println("Type: " + type);
        System.out.println("Boss? " + (isBoss ? "Yes" : "No"));
    }

    @Override
    public void attack(Character target) {
        int damage = isBoss ? 25 : 10;
        System.out.println(getName() + " slashes at " + target.getName() + " for " + damage + " damage!");
        target.setHealth(target.getHealth() - damage);
    }
}
````

Create a new class `Cleric.java`:
- Extend `Character`
- Implement **both** `Attacker` and `Healer`
- Add at least **two fields**, like:
  - `healingPower`
  - `damage`

Implement these methods:
- `display()` – show name, health, damage, healing power, etc.
- `attack(Character target)` – reduce target’s health
- `heal(Character target)` – increase target’s health

````python
%%writefile Cleric.java
public class Cleric extends Character implements Attacker, Healer {

    private int healingPower;
    private int damage;

    public Cleric(String name, int health, int healingPower, int damage) {
        super(name, health);
        this.healingPower = healingPower;
        this.damage = damage;
    }

    public int getHealingPower() {
        return healingPower;
    }
    public int getDamage() {
        return damage;
    }

    @Override
    public void display() {
        System.out.println("=== CLERIC ===");
        System.out.println("Name: " + getName());
        System.out.println("Health: " + getHealth());
        System.out.println("Healing Power: " + healingPower);
        System.out.println("Damage: " + damage);
    }

    @Override
    public void attack(Character target) {
        System.out.println(getName() + " smites " + target.getName() + " for " + damage + " damage!");
        target.setHealth(target.getHealth() - damage);
    }

    @Override
    public void heal(Character target) {
        System.out.println(getName() + " heals " + target.getName() + " for " + healingPower + " health!");
        target.setHealth(target.getHealth() + healingPower);
    }
}
````

````python
%%writefile Mage.java
public class Mage extends Character implements Attacker, Buffable{
    private int mana;
    private int buff;

    public Mage(String name, int health, int mana) {
        super(name, health);
        this.mana = mana;
    }

    public int getMana() {
        return mana;
    }

    @Override
    public void display() {
        System.out.println("=== MAGE ===");
        System.out.println("Name: " + getName());
        System.out.println("Health: " + getHealth());
        System.out.println("Mana Charges: " + mana);
      }

    @Override
    public void buff() {
        this.mana -= 1;
        this.buff += 1;
    }

    @Override
    public void attack(Character target) {
        int damage = 10 + buff*5;
        System.out.println(getName() + " zaps " + target.getName() + " for " + damage + " damage!");
        target.setHealth(target.getHealth() - damage);
    }
}
````

````python
%%writefile Rogue.java
public class Rogue extends Character implements Attacker {
  private boolean isHidden;

    public Rogue(String name, int health, boolean isHidden) {
        super(name, health);
        this.isHidden = isHidden;
    }

     @Override
    public void display() {
        System.out.println("=== Rogue ===");
        System.out.println("Name: " + getName());
        System.out.println("Health: " + getHealth());
        System.out.println("Hidden: " + isHidden);
      }

      public void hide() {
        this.isHidden = true;
      }

      public void reveal() {
        this.isHidden = false;
      }

      @Override
    public void attack(Character target) {
        int damage = 10;
        if (isHidden) {
            damage *= 2;
            isHidden = false;
        }
        System.out.println(getName() + " slashes " + target.getName() + " for " + damage + " damage!");
        target.setHealth(target.getHealth() - damage);
    }
}
````

````python
%%writefile Tank.java
public class Tank extends Character implements Attacker, Defender {
    private boolean defend;

    public Tank(String name, int health, boolean defend) {
        super(name, health);
        this.defend = defend;
    }

    public boolean isDefensive() {
        return defend;
    }

      @Override
    public void display() {
        System.out.println("=== Rogue ===");
        System.out.println("Name: " + getName());
        System.out.println("Health: " + getHealth());
        System.out.println("Defensive: " + defend);
      }

      @override
      public void defend() {
        this.defend = true;
        Tank.setHealth(Tank.getHealth() + 25);
      }

      @override
      public void undefend() {
        this.defend = false;
        Tank.setHealth(Tank.getHealth() - 25);
      }

      @Override
    public void attack(Character target) {
        int damage = 10;
        if (defend) {
            damage = 6;
        }
        System.out.println(getName() + " bashes " + target.getName() + " for " + damage + " damage!");
        target.setHealth(target.getHealth() - damage);
    }
}
````
